```python
""" LEVEL 1
    Task 1"""


"""Determine the top three most
 common cuisines in the dataset."""

data_top = data["Cuisines"].value_counts()
print(data_top.head(3))


"""Calculate
the percentage of restaurants that serve each of the top
cuisines"""

data_per = data_top / data_top.sum() * 100
# data per in percentage
data_per = data_per.reset_index()
data_per.columns = ["Cuisines", "Percentage"]
print(data_per.head(3))
```

```python
""" LEVEL 1
    Task 2"""

""" Task: City Analysis
 Identify the city with the highest number
 of restaurants in the dataset."""

data_city=data["City"].value_counts()
print(data_city.head(1))


"""Calculate the average rating for
 restaurants in each city."""

data_rating=data.groupby("City")["Aggregate rating"].mean()
print(data_rating)


"""Determine the city with the highest
 average rating"""

print(data_rating.sort_values(ascending=False).head(1))
```

```python
""" LEVEL 1
    Task 3"""


"""Task: Price Range Distribution
 Create a histogram or bar chart to
 visualize the distribution of price ranges
 among the restaurants with labels"""

plt.hist(data["Price range"])
plt.xlabel("Price Range")
plt.ylabel("Frequency")
plt.title("Price Range Distribution")
plt.show()


"""Calculate the percentage of restaurants
 in each price range category"""

data_price=data["Price range"].value_counts()
data_price=data_price/data_price.sum()*100
data_price=data_price.reset_index()
data_price.columns=["Price range","Percentage"]
print(data_price)
```

```python
""" LEVEL 1
    Task 4"""

""" Determine the percentage of restaurants
 that offer online delivery."""

data_delivery=data["Has Online delivery"].value_counts()
data_delivery=data_delivery/data_delivery.sum()*100
data_delivery=data_delivery.reset_index()
data_delivery.columns=["Has Online delivery","Percentage"]
print(data_delivery)


"""Compare the average ratings of restaurants
 with and without online delivery"""


data_delivery_rating=data.groupby("Has Online delivery")["Aggregate rating"].mean()
print(data_delivery_rating)


#  End of Level 1
```

```python
""" LEVEL 2
    Task 1"""

"""Task: Restaurant Ratings
 Analyze the distribution of aggregate
 ratings and determine the most common
 rating range."""

plt.figure(figsize=(6,4))
plt.hist(data['Aggregate rating'], bins=10, color='skyblue', edgecolor='black')
plt.xlabel('Aggregate Rating')
plt.ylabel('Frequency')
plt.title('Distribution of Restaurant Ratings')
plt.show()

common_rating=data["Aggregate rating"].value_counts()
print("Common Ratings",common_rating.head())

"""Calculate the Average number of votes
received by restaurants"""

data_votes=data["Votes"].mean()
print("Average number of votes received by restaurants",data_votes)
```

```python
""" LEVEL 2
    Task 2"""

"""Task: Cuisine Combination
 Identify the most common combinations of
 cuisines in the dataset."""

data_cuisine=data["Cuisines"].value_counts()
print(data_cuisine.head())

"""Determine if certain cuisine combinations
tend to have higher ratings"""

data_cuisine_rating=data.groupby("Cuisines")["Aggregate rating"].mean()
data_cuisine_rating.sort_values(ascending=False).head()
```

```python
""" LEVEL 2
    Task 3"""

""" Task: Geographic Analysis
 Plot the locations of restaurants on a
 map using longitude and latitude
 coordinates"""

plt.figure(figsize=(8,6))
plt.scatter(data["Longitude"], data["Latitude"], alpha=0.5, s=20, edgecolor='k')
plt.title("Geographic Distribution of Restaurants")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()


""" Identify any patterns or clusters of
 restaurants in specific areas"""
# patterns_restaurants=data.groupby("City")[["Longitude","Latitude"]].mean()
# patterns_restaurants

city_counts = data["City"].value_counts().head(5)
print("Top 5 Cities with the Most Restaurants:")
print(city_counts)
```

```python
""" LEVEL 2
    Task 4"""

"""Task: Restaurant Chains
 Identify if there are any restaurant chains
 present in the dataset"""
restaurant_chains=data["Restaurant Name"].value_counts()
print(restaurant_chains.head())

"""Analyze the ratings and popularity of
 different restaurant chains"""

ratings=data.groupby("Restaurant Name")["Aggregate rating"].agg(['count','mean']).sort_values('count',ascending=False)
ratings.head(10)


# End Level 2
```

```python
""" LEVEL 3
    Task 1"""

"""Task: Restaurant Reviews
 Analyze the text reviews to identify the most
 common positive and negative keywords."""

text_review=data["Rating text"].value_counts()
print(text_review)


"""Calculate the average length of reviews and
 explore if there is a relationship between
 review length and rating"""

# Calculate the length of the text in the 'Rating text' column
data['Rating text Length'] = data['Rating text'].str.len()
# Calculate the average length of rating text
avg_length = data['Rating text Length'].mean()
print(f"Average length of rating text: {avg_length}")

if 'Rating text' in data.columns:
    positive_words = ['Excellent', 'Very Good', 'Good']
    negative_words = ['Poor', 'Average']
    pos_count = data['Rating text'].isin(positive_words).sum()
    neg_count = data['Rating text'].isin(negative_words).sum()
    print("Positive keywords count:", pos_count)
    print("Negative keywords count:", neg_count)
```

```python
""" LEVEL 3
    Task 2"""

""" Identify the restaurants with the highest and
 lowest number of votes"""
restaurants_votes=data.sort_values(by="Votes",ascending=False)
top_votes=restaurants_votes[["Restaurant Name","City","Aggregate rating","Votes"]].head(2)
low_votes=restaurants_votes[["Restaurant Name","City","Aggregate rating","Votes"]].tail(2)
print(top_votes)
low_votes

corr_votes_rating = data['Votes'].corr(data['Aggregate rating'])
corr_votes_rating
```

```python
""" LEVEL 3
    Task 3"""

"""Task: Price Range vs. Online Delivery and
 Table Booking
 Analyze if there is a relationship between the
 price range and the availability of online
 delivery and table booking"""
data['Has Online delivery_numeric'] = data['Has Online delivery'].apply(lambda x: 1 if x == 'Yes' else 0)
data['Has Table booking_numeric'] = data['Has Table booking'].apply(lambda x: 1 if x == 'Yes' else 0)

relation_price_delivery = data.groupby("Price range")[['Has Online delivery_numeric', 'Has Table booking_numeric']].mean()*100
print(relation_price_delivery)


""" Determine if higher-priced restaurants are
 more likely to offer these services"""
relation_price_delivery.plot(kind='bar', figsize=(8,4))
plt.title('Online Delivery & Table Booking by Price Range (%)')
plt.ylabel('Percentage Offering Service')
plt.show()
```
